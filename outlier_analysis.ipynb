```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
import os
import pandas as pd
from sklearn.decomposition import PCA
fpc = "/home/admin123/Starspace/"+ "data/customer_rep.csv"
df = pd.read_csv(fpc)
rng = np.random.RandomState(42)
```

```python
req_cols = list(set(list(df.columns)) - set(list(["Customer"])))
X = df[req_cols].values
```

```python
# fit the model
clf = IsolationForest(max_samples=100,random_state=rng)
clf.fit(X)
ol_pred = clf.predict(X)
```

```python
is_outlier = (ol_pred == -1)
```

```python
ol_customers = df["Customer"][is_outlier]
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
os.chdir("/home/admin123/Starspace/")
fp = "/home/admin123/Starspace/"+ "data/Online_Retail.csv"
df_rd = pd.read_csv(fp)
df_rd = df_rd[-df_rd['InvoiceNo'].str.startswith("C")]
df_rd = df_rd[-df_rd['StockCode'].str.startswith("BANK")]
#req_cols = ["CustomerID", "InvoiceNo", "StockCode"]
df_rd = df_rd.dropna(how ='any')
df_rd["CustomerID"] = df_rd["CustomerID"].astype(int)
#df = df.sort_values(by=['CustomerID'])
df_rd["ItemTotal"] = df_rd["UnitPrice"] * df_rd["Quantity"]
```

```python
df_inv_summary = df_rd.groupby(["CustomerID","InvoiceNo"]).\
agg({'ItemTotal': ['sum'], 'Quantity':['sum'], 'StockCode':['count']}).reset_index()
```

```python
df_inv_summary.columns = [''.join(col).strip() for col in df_inv_summary.columns]

df_cust_summary = df_inv_summary.groupby('CustomerID').\
agg({'ItemTotalsum': ['sum'], 'InvoiceNo':['count']}).reset_index()
```
